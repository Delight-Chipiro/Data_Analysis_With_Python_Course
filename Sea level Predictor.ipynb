### Sea Level Predictor

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# 1 importing the data
df = pd.read_csv('epa-sea-level.csv')
df.head()
```

```python
# 2. Using seaborn and matplotlib for plotting the scatter plot and regression line
#x = df['Year']
#y = df['CSIRO Adjusted Sea Level']

sns.regplot(x='Year', y='CSIRO Adjusted Sea Level', data= df, line_kws={'color':'red'})
plt.title('Linear Regression with Best Fit Line')
plt.show()
#plt.scatter(df['Year'],df['CSIRO Adjusted Sea Level'])
```

```python
# 3,4,5 Predicting the sea level rise all the way to 2050 
future = df[df['Year'] >= 2000]
x = future['Year'].values.reshape(-1,1)
y = future['CSIRO Adjusted Sea Level'].values

model = LinearRegression()
model.fit(x,y)

years = np.arange(2000,2051).reshape(-1,1)
sea = model.predict(years)

plt.figure(figure=(10,6))

plt.scatter(future['Year'], future['CSIRO Adjusted Sea Level'], label='Data From 2000')
plt.plot(years,sea, label = 'Best fit Line (2000-2050)', color = 'red')

plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
plt.show()
```
